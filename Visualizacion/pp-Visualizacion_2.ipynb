```python
df=pd.read_csv('attacks2.csv', index_col=0)
df.drop(['Unnamed: 0.1','unnamed:_0'],axis=1, inplace=True)
```

```python
ataques_USA_ESP=df[df['country'].isin(['USA','SPAIN'])]
ataques_Sp=df[df['country'].isin(['SPAIN'])]
ataques_Us=df[df['country'].isin(['USA'])]
```

```python
fig,ax = plt.subplots(1,2, figsize =( 15,5))


ax[0].hist(ataques_Sp["type"], facecolor='red',edgecolor='black'  )

ax[1].hist(ataques_Us["type"], facecolor='blue',edgecolor='black')

ax[0].set_title("Tipos ataques en Espa√±a")
ax[1].set_title("Tipos ataques en USA")

ax[0].set_xlabel("Tipos de ataques")
ax[0].set_ylabel("Frecuencia")

ax[1].set_xlabel("Tipos de ataques")
ax[1].set_ylabel("Frecuencia")
 
plt.show()
```

```python


fig,ax = plt.subplots(2,1, figsize =( 15,10))

ax[0].hist(ataques_Sp["year"],bins=45,facecolor='yellow',edgecolor='black')

ax[1].hist(ataques_Us["year"],bins=45,facecolor='white',edgecolor='black')

ax[0].set_title("Ataques por a√±o en Espa√±a")
ax[1].set_title("Ataques por a√±o en en USA")

ax[0].set_xlabel("A√±o")
ax[0].set_ylabel("Frecuencia")

ax[1].set_xlabel("A√±o")
ax[1].set_ylabel("Frecuencia")


 
plt.show()
```

2. Cread un √∫nico gr√°fico donde junt√©is las gr√°ficas de los tipos de ataques en una sola. Deb√©is:

```python
y=ataques_Sp['type'].value_counts()
z=ataques_Us['type'].value_counts()
x=ataques_Sp['type'].unique()
w=ataques_Us['type'].unique()
```

```python
z
```

```python


plt.figure(figsize=(20,10))
plt.plot(y,x, label = 'Espa√±a', linewidth = 4, color = 'blue')
plt.plot(w,z, label = 'USA', linewidth = 4, color = 'green')

#Definir t√≠tulo y nombres de ejes
plt.title('Tipo de ataques en Espa√±a y USA')
plt.ylabel('Tipo de Ataque')
plt.xlabel('Frecuencia')
plt.xlim()
plt.ylim()

#Mostrar leyenda, cuadr√≠cula y figura
plt.legend()
plt.grid()
plt.show()
```
