- ¿Cuál es el porcentaje de valores nulos en cada columna?

- Sacad los valores únicos y su frecuencia de las variables numéricas. Almacenad los resultados en un diccionario. Uno para cada variable.

```python
dic_unicos ={} #Creamos un diccionario con variables numericas
for i in df:
    
    if df.dtypes[i] == 'object':
        pass
    else:
         dic_unicos.update({i:(df[i].value_counts())})
```

```python
dic_unicos.keys() #Verificamos que columnas trajo
```

```python
# convertimos el diccionario en tres uno por cada variable numerica como pedia la consigna.

dic_unnamed = dict(dic_unicos['Unnamed: 0']) 
dic_year = dict(dic_unicos['Year'])
dic_original = dict(dic_unicos['original order'])
```
